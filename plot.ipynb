```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'num_amount': ['100k', '200k', '300k', '400k', '500k', '600k', '700k', '800k', '900k', '1000k'],
                   'avg_time':   [ 0.025, 0.049 , 0.055 , 0.087 , 0.093 , 0.111 , 0.134 , 0.148 , 0.170 , 0.184  ]})
sns.lineplot(data=df, x="num_amount", y="avg_time", color = 'pink')
plt.title('График зависимости времени выполнения функции _max от кол-ва чисел в файле', fontsize=10)
plt.figure(figsize=(10,5), dpi=600)
```
